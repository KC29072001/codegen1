natural lang module

```python


def preprocess_input(text):
    tokens = word_tokenize(text)
    tokens = [word.lower() for word in tokens if word.isalnum() and word.lower() not in stopwords.words('english')]
    return tokens
```

Code gen. module

```python
# Set the API key using the OpenAI class

client = OpenAI(api_key='YOUR_API_KEY')

# text = data[0].page_content

def generate_code(text):
    # Prompting the GPT-3 model with the natural language description
    prompt = f"Translate the following natural language description into code:\n\n{text}\n\nCode:"

    # Requesting code generation from GPT-3
    response = client.chat.completions.create(
        # engine="text-davinci-002",
        messages = [ # Change the prompt parameter to messages parameter
    {'role': 'user', 'content': text}],
       temperature = 0,  
        # prompt=prompt,
        model="gpt-3.5-turbo",
        max_tokens=500
    )

    # Extracting the generated code from the response
    print(response.choices[0].message.content)






```

```python
text1 = data1[0].page_content
text2 = data2[0].page_content
text3 = data3[0].page_content


preprocessed_texts = []

# Iterate over the texts
for text in [text1, text2, text3]:
    # Call the preprocess_input function on each text and store the result
    preprocessed_text = preprocess_input(text)
    listToStr = ' '.join(map(str,preprocessed_text))
    generate_code(listToStr)
    
    # Append the preprocessed text to the array
#     preprocessed_texts.append(preprocessed_text)
# print(preprocessed_texts)
```
